```python
import requests
import pandas as pd
import time

API_KEY = "26ZPQ4V-P2SM9Q8-MWP0GFZ-NK9DMFN"
BASE_URL = "https://api.kinopoisk.dev/v1.4/movie/search"
headers = {"X-API-KEY": API_KEY}


movies = [
    "Побег из Шоушенка",
    "Крестный отец",
    "Темный рыцарь",
    "Крестный отец 2",
    "12 разгневанных мужчин",
    "Властелин колец: Возвращение короля",
    "Список Шиндлера",
    "Криминальное чтиво",
    "Властелин колец: Братство Кольца",
    "Хороший, плохой, злой",
    "Форрест Гамп",
    "Властелин колец: Две крепости",
    "Бойцовский клуб",
    "Начало",
    "Звёздные войны. Эпизод V: Империя наносит ответный удар",
    "Матрица",
    "Славные парни",
    "Интерстеллар",
    "Пролетая над гнездом кукушки",
    "Семь"
]


movies_info = []

for title in movies:
    params = {"query": title, "limit": 1}
    response = requests.get(BASE_URL, headers=headers, params=params)

    if response.status_code == 200:
        data = response.json()
        docs = data.get("docs", [])
        if docs:
            film = docs[0]
            movies_info.append({
                "Title": film.get("name"),
                "Alternative Title": film.get("alternativeName"),
                "Year": film.get("year"),
                "Length (min)": film.get("movieLength"),
                "Kinopoisk Rating": film.get("rating", {}).get("kp"),
                "IMDb Rating": film.get("rating", {}).get("imdb"),
                "Genres": ", ".join([g["name"] for g in film.get("genres", [])]),
                "Countries": ", ".join([c["name"] for c in film.get("countries", [])])
            })



df = pd.DataFrame(movies_info)

df



```

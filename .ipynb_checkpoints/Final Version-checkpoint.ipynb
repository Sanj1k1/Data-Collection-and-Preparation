```python
#Web Scrapping HTML page wiki
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://ru.wikipedia.org/wiki/250_%D0%BB%D1%83%D1%87%D1%88%D0%B8%D1%85_%D1%84%D0%B8%D0%BB%D1%8C%D0%BC%D0%BE%D0%B2_%D0%BF%D0%BE_%D0%B2%D0%B5%D1%80%D1%81%D0%B8%D0%B8_IMDb"
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/141.0.0.0 Safari/537.36"}
response = requests.get(url,headers = headers)
print("Status:", response.status_code)
```

```python
soup = BeautifulSoup(response.text, "html.parser")
```

```python
table = soup.find("table",{"class": "sortable"})
```

```python
headers = [th.get_text(strip=True) for th in table.find_all("th")]
print(headers)
```

```python
data = []
for row in table.find("tbody").find_all("tr"):
    cells = [td.get_text(strip=True) for td in row.find_all("td")]
    data.append(cells)
```

```python
df2 = pd.DataFrame(data[1:21],columns = headers)
df2.head(20)
```

```python
#API kinopoisk.dev to parse data
API_KEY = "ZWZ43H6-6CVMZ54-MVS19B7-M7E0V8G"
BASE_URL = "https://api.kinopoisk.dev/v1.4/movie/search"
headers = {"X-API-KEY": API_KEY}

movies = [
    'Побег из Шоушенка',
    'Крестный отец',
    'Темный рыцарь',
    'Крестный отец 2',
    '12 разгневанных мужчин',
    'Властелин колец: Возвращение короля',
    'Список Шиндлера',
    'Властелин колец: Братство Кольца',
    'Криминальное чтиво',
    'Хороший, плохой, злой',
    'Форрест Гамп',
    'Властелин колец: Две крепости',
    'Бойцовский клуб',
    'Начало',
    'Звездные войны. Эпизод V: Империя наносит ответный удар',
    'Матрица',
    'Славные парни',
    'Интерстеллар',
    'Пролетая над гнездом кукушки',
    'Семь',
]


movies_info = []

for title in movies:
    params = {"query": title, "limit": 1}
    response = requests.get(BASE_URL, headers=headers, params=params)

    if response.status_code == 200:
        data = response.json()
        docs = data.get("docs", [])
        if docs:
            film = docs[0]
            movies_info.append({
                "Название фильма": film.get("name"),
                "Альтернативное название фильма": film.get("alternativeName"),
                "Год": film.get("year"),
                "Длительность (мин)": film.get("movieLength"),
                "Рейтинг Кинопоиск": film.get("rating", {}).get("kp"),
                "Рейтинг IMDb": film.get("rating", {}).get("imdb"),
                "Жанры по версии Кинопоиск": ", ".join([g["name"] for g in film.get("genres", [])]),
                "Страна(-ы)": ", ".join([c["name"] for c in film.get("countries", [])])
            })



df = pd.DataFrame(movies_info)

df
```

```python
df["Год"] = df["Год"].astype(int)
df2["Год"] = df2["Год"].astype(int)
```

```python
#Merge datas
df2["Название фильма"] = df["Название фильма"].str.replace("ё", "е")
merged = pd.merge(df, df2,left_on=["Название фильма"], right_on=["Название фильма"],how="inner")
#Убралли не состыкующие страны и добавили в один общий
merged["Страна(-ы)"] = merged["Страна(-ы)_x"]
merged = merged.drop(columns=["Страна(-ы)_x", "Страна(-ы)_y"])
idx = 2 
merged.insert(loc=idx,column="Год",value=merged["Год_x"])
merged = merged.drop(columns=["Год_x", "Год_y"])

merged
```

```python
#Visualisation and Quality
#scatter plot 
import matplotlib.pyplot as plt
correlation = merged['Рейтинг Кинопоиск'].corr(df['Рейтинг IMDb'])
print(f"Корреляция между рейтингами Кинопоиска и IMDb: {correlation:.3f}")
plt.figure(figsize=(8, 6))
plt.scatter(merged["Рейтинг IMDb"], merged["Рейтинг Кинопоиск"],alpha=0.7,edgecolors="black")
min_r = min(df['Рейтинг Кинопоиск'].min(), df['Рейтинг IMDb'].min())
max_r = max(df['Рейтинг Кинопоиск'].max(), df['Рейтинг IMDb'].max())
plt.plot([min_r, max_r], [min_r, max_r], 'r--', alpha=0.5, label='Идеальная корреляция')
plt.title("Сравнение рейтингов IMDb и Кинопоиск")
plt.xlabel("Рейтинг IMDb")
plt.ylabel("Рейтинг Кинопоиск")
plt.grid(True)
plt.show()
```

```python
import seaborn as sns
sns.scatterplot(x='Рейтинг IMDb', y='Рейтинг Кинопоиск', data= merged)
print(f"Корреляция между рейтингами Кинопоиска и IMDb: {correlation:.3f}")
plt.title("Сравнение рейтингов IMDb и Кинопоиск")
plt.plot([min_r, max_r], [min_r, max_r], 'r--', alpha=0.5, label='Идеальная корреляция')
plt.show()
```

```python
#bar ТОП 10 фильмов по рейтингу
df_imdb = merged.copy()
df_imdb = df_imdb.sort_values(by="Рейтинг IMDb", ascending=False)
plt.figure(figsize=(10,5))
plt.bar(
    df_imdb["Название фильма"].head(10),   
    df_imdb["Рейтинг IMDb"].head(10),
    color='skyblue',
    edgecolor='black',
    width=0.5
)

plt.title("Рейтинг и IMDB top 10 фильмов")
plt.ylabel("Рейтинг IMDB")
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
#распределения рейтингов IMDb и Кинопоиска.
plt.figure(figsize=(8, 5))
sns.histplot(merged['Рейтинг IMDb'], bins=10, kde=True, color='skyblue',alpha=0.6)
sns.histplot(merged['Рейтинг Кинопоиск'], bins=10, kde=True, color='red',alpha=0.3)
plt.title('Распределение рейтингов IMDb')
plt.xlabel('Рейтинг IMDb')
plt.ylabel('Количество фильмов')
plt.show()
```

```python
#рейтинг по годам
plt.figure(figsize=(10, 6))
sns.lineplot(data=merged, x='Год', y='Рейтинг IMDb')
sns.lineplot(data=merged, x='Год', y='Рейтинг Кинопоиск')
plt.axhline(merged['Рейтинг IMDb'].mean(), 
            color='green', linestyle='-', 
            alpha=0.7, 
            label=f'Средний IMDb: {merged["Рейтинг IMDb"].mean():.2f}'
           )
plt.axhline(merged['Рейтинг Кинопоиск'].mean(), 
            color='red',
            linestyle='-.', 
            alpha=0.7,
            label=f'Средний Кинопоиск: {merged["Рейтинг Кинопоиск"].mean():.2f}'
           )
plt.title('Изменение рейтингов IMDb и Кинопоиска по годам')
plt.xlabel('Год')
plt.ylabel('Рейтинг')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#длительность по годам
sns.lineplot(data=merged, x='Год', y='Длительность (мин)')
mean_duration = merged['Длительность (мин)'].mean()
plt.axhline(mean_duration, color='red', linestyle=':', label=f'({mean_duration} минут в среднем)')
plt.title('Изменение длительности фильмов по годам')
plt.xlabel('Год')
plt.ylabel('Длительность (мин)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Analysis
#Ratings
mean_imbd_rating = merged['Рейтинг IMDb'].mean()
mean_kp_rating = merged['Рейтинг Кинопоиск'].mean()
max_kp_rating =  merged['Рейтинг Кинопоиск'].max()
min_kp_rating =  merged['Рейтинг Кинопоиск'].min()
max_imbd_rating = merged['Рейтинг IMDb'].max()
min_imbd_rating = merged['Рейтинг IMDb'].min()
avg_rating_by_country = merged.groupby('Страна(-ы)')[['Рейтинг IMDb', 'Рейтинг Кинопоиск']].mean().round(2)

#Durations
mean_duration = merged['Длительность (мин)'].mean()
max_duration = merged["Длительность (мин)"].max()
min_duration = merged["Длительность (мин)"].min()
name_film_max_dur = merged.loc[merged["Длительность (мин)"]==max_duration,"Название фильма"].values[0]
name_film_min_dur = merged.loc[merged["Длительность (мин)"]==min_duration,"Название фильма"].values[0]

#Correlations 
correlation_ratings = merged['Рейтинг Кинопоиск'].corr(df['Рейтинг IMDb'])
correlation_kprating_dur = merged['Длительность (мин)'].corr(merged['Рейтинг Кинопоиск'])
correleation_imdbrating_dur = merged['Длительность (мин)'].corr(merged['Рейтинг IMDb'])

print(f"Средний Рейтинг IMDb: {mean_imbd_rating}")
print(f"Средний Рейтинг Кинопоиск: {mean_kp_rating}")
print(f"Максимальный и Минимальный Рейтинг IMDb: {max_imbd_rating} , {min_imbd_rating}")
print(f"Максимальный и Минимальный Рейтинг Кинопоиск: {max_kp_rating} , {min_kp_rating}")
print()
print(avg_rating_by_country)
print()
print(f"Средняя длительность фильмов: {mean_duration}")
print(f"Максимальная длителность фильма: {max_duration}")
print(f"Минимальная длительность фильма: {min_duration}")
print(f"Самый длинный фильм: {name_film_max_dur}")
print(f"Самый маленький фильм: {name_film_min_dur}")
print()
print(f"Корреляция рейтингом Кинопоиск и рейтингом IMDb: {correlation_ratings}" )
print(f"Корреляция длительности с рейтингом Кинопоиск: {correlation_kprating_dur}")
print(f"Корреляция длительности с рейтингом IMDb: {correleation_imdbrating_dur}")
```
